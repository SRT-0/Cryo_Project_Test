```python
# @title 3. Parameter set up

block_size = 64                            # @param{type : "integer"}
# Number of blocks take (how many sample for each micrograph)
blocks_num = 30                            # @param{type : "integer"}
particle_filter_block_size = 128           # @param{type : "integer"}
# Micrograph edge exclusion (by pixels)
border_size = 120                          # @param{type : "integer"}
min_block_distance = 32                    # @param{type : "integer"}
max_background_particles_distance = 150    # @param{type : "integer"}

# Minimum distance required for 2 different blocks
# Maximum number of SNR pairs to compute per micrograph

save_plots = True                          # @param{type : "boolean"}

flip_y_idx = True                          # @param{type : "boolean"}

# @markdown ---
# @markdown if u want to use all micrograph for computing SNR then check ✔ the all_micrograph_for_SNR
num_micrograph_for_SNR = 35              # @param{type : "integer"}
all_micrograph_for_SNR = False           # @param{type : "boolean"}

if all_micrograph_for_SNR == True:
    num_micrograph_for_SNR = None

random_state = 42                          # @param{type : "integer"}


EDGE_CLEARANCE_FACTOR = 1.5

@dataclass
class SNR_config:
    """
    Configuration for the SNR calculation process.
    """
    # PATH
    MICROGRAPH_DIR: Path = MICROGRAPH_DIR
    PARTICLE_COOR_DIR: Path = PARTICLE_COOR_DIR
    MASK_DIR: Path = MASK_DIR
    OUTPUT_DIR: Path = OUTPUT_DIR

    # Parameters
    block_size: int = block_size
    blocks_num: int = blocks_num
    PF_block_size: int = particle_filter_block_size
    random_state: Optional[int] = random_state
    border_size: int = border_size
    min_block_distance: int = min_block_distance
    max_background_particles_distance: int = max_background_particles_distance

    SNR_list: List[float] = field(default_factory=list)
    mrc_list: List[float] = field(default_factory=list)
    mrc_list_picked: List[float] = field(default_factory=list)

    num_micrograph_for_SNR: Optional[int] = num_micrograph_for_SNR

    do_plots: bool = save_plots

    flip_y: bool = flip_y_idx

    def __post_init__(self):
        """Initializes random seed for reproducibility and ensures output directory exists."""
        if self.random_state is not None:
            np.random.seed(self.random_state)
            random.seed(self.random_state)

        # Ensure output directory exists before writing results
        self.OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

        # Reset SNR_list if it was somehow initialized with values (though field(default_factory=list) handles this)
        if not hasattr(self, 'SNR_list'):
            self.SNR_list = []
            self.mrc_list = []
            self.mrc_list_picked = []
```

```python
# @title 4. Functions setup

# --- Utility Functions (Data Loading/Saving) ---
def load_mrcfile(cfg: SNR_config) -> List[Tuple[np.ndarray, str]]:
    """
    Loads all micrograph data by recursively searching for *.mrc files
    within the MICROGRAPH_DIR.
    """
    mrc_files = list(cfg.MICROGRAPH_DIR.rglob('*.mrc'))
    print(f"Found {len(mrc_files)} .mrc files in {cfg.MICROGRAPH_DIR}.")
    micrographs = []

    if not mrc_files:
        raise FileNotFoundError(
            f"No .mrc files found in {cfg.MICROGRAPH_DIR} or its subdirectories. Cannot proceed."
        )

    for mrc_path in mrc_files:
        mrc_filename = mrc_path.name

        try:
            with mrcfile.open(mrc_path) as mrc:
                img_data = mrc.data.astype(np.float64)

            print(f"Successfully loaded {mrc_filename} with shape {img_data.shape}.")
            micrographs.append((img_data, mrc_filename))

        except Exception as e:
            print(f"WARNING: Could not load or read MRC file {mrc_path}. Skipping. Error: {e}")
            continue

    return micrographs

def load_particle_coor(cfg: SNR_config, coor_path: Path, img_height: int) -> np.ndarray:
    """
    Loads particle coordinates for a micrograph from a CSV file.
    """
    # print(f"Loading particle coordinates from {coor_path}...")
    try:
        with open(coor_path, 'r', newline='') as csvfile:
            reader = csv.reader(csvfile)
            header = next(reader)

            try:
                x_idx = header.index('X-Coordinate')
                y_idx = header.index('Y-Coordinate')
            except ValueError:
                raise ValueError("CSV file header must contain 'X-Coordinate' and 'Y-Coordinate' columns.")

            coords = []
            for row in reader:
                try:
                    x = float(row[x_idx])
                    y = float(row[y_idx])
                    coords.append((x, y))
                except (ValueError, IndexError):
                    print(f"Warning: Skipping invalid row in {coor_path}: {row}")
                    continue

        if not coords:
            print("WARNING: Coordinate file was read but contained no valid particle data.")
            return np.array([])

        coords = np.array(coords, dtype=np.float64)

        if cfg.flip_y:
            coords[:, 1] = img_height - 1 - coords[:, 1]

        return coords

    except FileNotFoundError:
        print(f"ERROR: Coordinate file not found at {coor_path}. Returning empty array.")
        return np.array([])
    except Exception as e:
        print(f"ERROR: Failed to load coordinates from {coor_path}. Error: {e}")
        return np.array([])

def SNR_csv_write_micrograph_data(cfg: SNR_config, mrc_filename: str,
                                  mean_micrograph_snr: float, total_pairs_N: int,
                                  file_name: str = "micrograph_snr_results.csv"):
    """
    Writes the single mean SNR result for a micrograph to a CSV file in the OUTPUT_DIR.
    """
    output_path = cfg.OUTPUT_DIR / file_name
    file_exists = output_path.exists()

    try:
        with open(output_path, mode='a', newline='') as file:
            writer = csv.writer(file)

            if not file_exists:
                writer.writerow([
                    "Micrograph_Filename",
                    "Total_Pairs_N",
                    "Mean_SNR_Micrograph"
                ])

            writer.writerow([
                mrc_filename,
                total_pairs_N,
                f"{mean_micrograph_snr:.6f}"
            ])

    except Exception as e:
        print(f"ERROR: Failed to write to CSV file {output_path}. Error: {e}")


# --- Core SNR Calculation Functions ---
def _extract_block(img_data: np.ndarray, x: int, y: int, size: int) -> np.ndarray:
    """
    Extracts a square block of 'size' x 'size' pixels centered at (x, y).
    """
    h, w = img_data.shape

    if size % 2 == 0: # Even
        half_size = size // 2
        x_start = max(0, x - (half_size - 1))
        x_end = min(w, x + half_size)
        y_start = max(0, y - half_size)
        y_end = min(h, y + (half_size - 1))
    else: # Odd
        size_adj = size - 1
        half_size = size_adj // 2
        x_start = max(0, x - half_size)
        x_end = min(w, x + half_size)
        y_start = max(0, y - half_size)
        y_end = min(h, y + half_size)

    block = img_data[y_start:y_end, x_start:x_end]
    return block

def SNR_pre_computation(particle_region: np.ndarray, background_region: np.ndarray) -> float:
    """
    SNR: log10( Var(Particle - Mean(Background)) / Var(Background) )
    """
    v_b = np.var(background_region)
    mu_b = np.mean(background_region)

    if v_b <= 1e-12:
        return 0.0

    s = particle_region - mu_b
    v_s = np.var(s)

    snr_i = math.log10(v_s / v_b)
    return snr_i

# --- Helper for Distance Transform ---
def _calculate_distance_map(pf_binary_mask: np.ndarray) -> np.ndarray:
    """
    Calculates the minimum Euclidean distance from every background pixel (0)
    to the nearest particle pixel (1) in the mask.
    """
    particle_pixels_y, particle_pixels_x = np.where(pf_binary_mask == 1)
    particle_pixel_coords = np.vstack((particle_pixels_x, particle_pixels_y)).T

    if particle_pixel_coords.size == 0:
        return np.full(pf_binary_mask.shape, np.inf, dtype=np.float32)

    particle_tree = KDTree(particle_pixel_coords)

    h, w = pf_binary_mask.shape
    y_coords, x_coords = np.mgrid[0:h, 0:w]
    all_pixel_coords = np.vstack((x_coords.ravel(), y_coords.ravel())).T

    distances_flat, _ = particle_tree.query(all_pixel_coords, k=1)
    distance_map = distances_flat.reshape(h, w)
    return distance_map

def plot_selected_regions(
    cfg: SNR_config,
    img_data: np.ndarray,
    img_filename: str,
    particle_centers: list,
    background_centers: list
):
    """
    Generates and saves a plot showing the selected particle and background blocks.
    """
    output_path = cfg.OUTPUT_DIR / "plots_selected_for_SNR" / f"{Path(img_filename).stem}_blocks.png"
    output_path.parent.mkdir(parents=True, exist_ok=True)

    plt.figure(figsize=(12, 12))
    plt.imshow(img_data, cmap='gray')
    ax = plt.gca()

    block_size = cfg.block_size
    half_block = block_size / 2

    # Plotting particle blocks (red)
    for x, y in particle_centers:
        rect = Rectangle(
            (x - half_block, y - half_block),
            block_size,
            block_size,
            linewidth=1.5,
            edgecolor='r',
            facecolor='none',
            alpha=0.8
        )
        ax.add_patch(rect)

    # Plotting background blocks (blue)
    for x, y in background_centers:
        rect = Rectangle(
            (x - half_block, y - half_block),
            block_size,
            block_size,
            linewidth=1.5,
            edgecolor='b',
            facecolor='none',
            alpha=0.8
        )
        ax.add_patch(rect)

    plt.title(f"Selected Blocks for {img_filename}", fontsize=16)
    plt.xlabel("X-coordinate")
    plt.ylabel("Y-coordinate")
    plt.tight_layout()
    plt.savefig(output_path, dpi=300)
    plt.close()

# --- Version 1: Block-Based (Smart Distance Sampling) Logic ---
def _remove_particle_blocks(cfg: SNR_config, np_img_data: np.ndarray, particles_coor: np.ndarray) -> Tuple[np.ndarray, np.ndarray]:
    """
    Creates a binary mask (1=particle, 0=background) for the Distance Transform,
    and the content mask (NaNs) for content exclusion.
    """
    h, w = np_img_data.shape
    pf_binary_mask = np.zeros((h, w), dtype=np.int8)
    pf_size = cfg.PF_block_size

    if pf_size % 2 == 0:
        pf_half = pf_size // 2
        pf_half_adj = pf_half - 1
    else:
        pf_size_adj = pf_size - 1
        pf_half = pf_size_adj // 2
        pf_half_adj = pf_half

    masked_img_content = np_img_data.copy()
    mask_value = np.nan

    for x, y in particles_coor:
        x, y = int(x), int(y)
        x_start = max(0, x - pf_half_adj)
        x_end = min(w, x + pf_half)
        y_start = max(0, y - pf_half)
        y_end = min(h, y + pf_half_adj)

        pf_binary_mask[y_start:y_end, x_start:x_end] = 1
        masked_img_content[y_start:y_end, x_start:x_end] = mask_value

    return masked_img_content, pf_binary_mask

def _pick_background_coords(cfg: SNR_config, masked_img_content: np.ndarray, pf_binary_mask: np.ndarray) -> List[Tuple[int, int]]:
    """
    Background selection blocks.
    """
    h, w = masked_img_content.shape
    distance_map = _calculate_distance_map(pf_binary_mask)

    size = cfg.block_size
    if size % 2 == 0:
        half_floor = size // 2 - 1
    else:
        half_floor = size // 2

    min_x = cfg.border_size + half_floor
    max_x = w - cfg.border_size - half_floor
    min_y = cfg.border_size + half_floor
    max_y = h - cfg.border_size - half_floor

    if max_x <= min_x or max_y <= min_y:
        return []

    min_required_clearance = cfg.block_size * EDGE_CLEARANCE_FACTOR
    MAX_MASK_CONTAMINATION = 0.03
    clean_centers = []
    SAMPLE_STEP = 8

    for y in range(min_y, max_y, SAMPLE_STEP):
        for x in range(min_x, max_x, SAMPLE_STEP):
            if (distance_map[y, x] > min_required_clearance and
                distance_map[y, x] <= cfg.max_background_particles_distance):

                mask_block = _extract_block(pf_binary_mask, x, y, cfg.block_size)

                if np.mean(mask_block) < MAX_MASK_CONTAMINATION:
                    clean_centers.append((x, y))

    if not clean_centers:
        return []

    final_coords = []
    rng = np.random.default_rng(cfg.random_state)
    rng.shuffle(clean_centers)
    min_center_dist = cfg.block_size + cfg.min_block_distance

    for x, y in clean_centers:
        if len(final_coords) >= cfg.blocks_num:
            break

        is_too_close = False
        for px, py in final_coords:
            dist_x = abs(x - px)
            dist_y = abs(y - py)
            if dist_x < min_center_dist and dist_y < min_center_dist:
                is_too_close = True
                break

        if not is_too_close:
            final_coords.append((x, y))

    return final_coords

def SNR_within_blocks_block_based(cfg: SNR_config, np_img_data: np.ndarray, particles_coor: np.ndarray, img_filename: str) -> Tuple[float, int]:
    """
    Computes DNR by block
    """
    num_to_sample = cfg.blocks_num
    rng = np.random.default_rng()
    num_particles_available = len(particles_coor)
    current_num_sets = min(num_to_sample, num_particles_available)

    if current_num_sets == 0:
        return 0.0, 0

    indices = rng.choice(num_particles_available, size=current_num_sets, replace=False)
    particle_centers = particles_coor[indices]

    masked_img_content, pf_binary_mask = _remove_particle_blocks(cfg, np_img_data, particles_coor)
    background_centers_all = _pick_background_coords(cfg, masked_img_content, pf_binary_mask)

    current_num_sets = min(current_num_sets, len(background_centers_all))
    if current_num_sets == 0:
        return 0.0, 0

    background_centers = random.sample(background_centers_all, current_num_sets)
    particle_centers = particle_centers[:current_num_sets]

    if cfg.do_plots:
        plot_selected_regions(cfg, np_img_data, img_filename, particle_centers, background_centers)

    snr_list = []
    for i in range(current_num_sets):
        px, py = particle_centers[i]
        particle_region = _extract_block(np_img_data, int(px), int(py), cfg.block_size)

        bx, by = background_centers[i]
        background_region = _extract_block(np_img_data, bx, by, cfg.block_size)

        if particle_region.size > 0 and not np.all(particle_region == 0) and \
           background_region.size > 0 and not np.all(background_region == 0):
            snr_i = SNR_pre_computation(particle_region, background_region)
            snr_list.append(snr_i)

    N = len(snr_list)
    snr_val = 10 * np.mean(snr_list) if N > 0 else 0.0

    return snr_val, N

# --- Multiprocessing Wrapper Function ---
def _process_single_micrograph(data_tuple: Tuple[np.ndarray, str], config: SNR_config):
    """
    This function will be run by a worker process in the multiprocessing pool.
    It performs the SNR calculation for a single micrograph and returns the result.
    """
    img_data, img_filename = data_tuple

    file_stem = Path(img_filename).stem

    try:
        h, w = img_data.shape
        coor_path = config.PARTICLE_COOR_DIR / f"{file_stem}.csv"
        particles_coor = load_particle_coor(config, coor_path, h)

        micrograph_mean_snr, N = SNR_within_blocks_block_based(config, img_data, particles_coor, img_filename)

        return (micrograph_mean_snr, N, img_filename)

    except Exception as e:
        # A failed process will return (0.0, 0, filename) to be handled gracefully
        print(f"CRITICAL ERROR during SNR calculation for {img_filename}: {e}.")
        return (0.0, 0, img_filename)

# --- Main Class and Orchestration ---
class SNRCalculator:
    """
    SNR Class for user to call.
    """
    def __init__(self, config: SNR_config = SNR_config):
        config.mrc_list = []
        config.mrc_list_picked = []
        if config.blocks_num < 1:
            raise ValueError(f"blocks_num must be at least 1 for mean calculation. Got {config.blocks_num}")

    def calculate_final_stats(self, config: SNR_config = SNR_config):
        """
        Computes, prints, and records final statistics on the collected
        micrograph mean SNRs.
        """
        samples = np.array(config.SNR_list)
        N = len(samples)

        if N < 2:
            output = f"\n--- Final Statistics (Skipped) ---\n" \
                     f"Need at least 2 micrograph mean SNRs to compute CI/Box Plot data. Found N={N}.\n" \
                     f"---------------------------------"
            print(output)
            return

        output_lines = []
        output_lines.append("\n" + "=" * 40)
        output_lines.append("Final Statistical Summary (Across all Micrographs)")
        output_lines.append("=" * 40)

        mean_all = np.mean(samples)
        std_err = stats.sem(samples)
        degrees_freedom = N - 1

        if std_err == 0:
            ci_lower = ci_upper = mean_all
        else:
            ci_margin = std_err * t.ppf(0.975, degrees_freedom)
            ci_lower = mean_all - ci_margin
            ci_upper = mean_all + ci_margin

        output_lines.append(f"Total Micrographs (N): {N}")
        output_lines.append(f"Overall Mean SNR: {mean_all:.6f}")
        output_lines.append(f"95% Confidence Interval: ({ci_lower:.4f}, {ci_upper:.4f})")
        output_lines.append("\n[Visualization Data for Box/Density Plots]")
        output_lines.append("Sample distribution statistics (using Micrograph Mean SNRs as samples):")
        output_lines.append(f"  Median: {np.median(samples):.4f}")
        output_lines.append(f"  25th Percentile (Q1): {np.percentile(samples, 25):.4f}")
        output_lines.append(f"  75th Percentile (Q3): {np.percentile(samples, 75):.4f}")
        output_lines.append("  Full Sample List (for plotting density/box plot):")

        if N > 10:
            sample_str = f"    {samples[:5].tolist()} ... (showing first 5)"
        else:
            sample_str = f"    {samples.tolist()}"
        output_lines.append(sample_str)
        output_lines.append("---------------------------------")
        output_lines.append("\n" + "=" * 60)
        output_lines.append(f"Overall Mean Micrograph SNR: {mean_all:.6f}")
        output_lines.append("=" * 60)

        final_output_str = "\n".join(output_lines)
        print(final_output_str)

        output_path_txt = config.OUTPUT_DIR / "SNR_rst.txt"
        try:
            with open(output_path_txt, 'w') as f:
                f.write(final_output_str)
            print(f"Final summary written to {output_path_txt}")
        except Exception as e:
            print(f"ERROR: Failed to write overall summary to {output_path_txt}. Error: {e}")

    def mean_SNR_within_micrographs(self, config: SNR_config = SNR_config) -> float:
        """
        Orchestrates the parallel processing of micrographs using a Pool and tqdm.
        """
        data_list = load_mrcfile(config)
        n_total_micrographs = len(data_list)
        if n_total_micrographs == 0:
            print("No micrograph data loaded. Exiting.")
            return 0.0

        if config.num_micrograph_for_SNR is not None and config.num_micrograph_for_SNR < n_total_micrographs:
            num_to_sample = config.num_micrograph_for_SNR
            rng = random.Random(config.random_state)
            sampled_micrographs = rng.sample(data_list, num_to_sample)
            print(f"Randomly selected {num_to_sample} micrographs for SNR calculation.")
        else:
            num_to_sample = n_total_micrographs
            sampled_micrographs = data_list
            print(f"Using all {n_total_micrographs} micrographs for SNR calculation.")

        print(f"\nProcessing {num_to_sample} micrographs using {cpu_count()} CPU cores...")

        process_func = partial(_process_single_micrograph, config=config)

        config.SNR_list = []

        with Pool(processes=cpu_count()) as pool:
            results_iterator = pool.imap(process_func, sampled_micrographs)

            for micrograph_mean_snr, N, img_filename in tqdm(results_iterator, total=len(sampled_micrographs), desc="Calculating SNR"):
                if N > 0:
                    SNR_csv_write_micrograph_data(
                        cfg=config,
                        mrc_filename=img_filename,
                        mean_micrograph_snr=micrograph_mean_snr,
                        total_pairs_N=N
                    )
                    config.SNR_list.append(micrograph_mean_snr)

        self.calculate_final_stats(config)

        if not config.SNR_list:
            mean_mSNR = 0.0
        else:
            mean_mSNR = np.mean(config.SNR_list)

        return mean_mSNR
```
