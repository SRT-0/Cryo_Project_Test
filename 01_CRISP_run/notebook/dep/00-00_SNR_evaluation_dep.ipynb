```python
# @title 4. Functions setup

# --- Utility Functions (Data Loading/Saving) ---
def load_mrcfile(cfg: SNR_config) -> List[Tuple[np.ndarray, str]]:
    """
    Loads all micrograph data by recursively searching for *.mrc files
    within the MICROGRAPH_DIR.
    """

    mrc_files = list(cfg.MICROGRAPH_DIR.rglob('*.mrc'))
    print(f"Found {len(mrc_files)} .mrc files in {cfg.MICROGRAPH_DIR}.")
    micrographs = []

    if not mrc_files:
        raise FileNotFoundError(
            f"No .mrc files found in {cfg.MICROGRAPH_DIR} or its subdirectories. Cannot proceed."
        )

    for mrc_path in mrc_files:
        mrc_filename = mrc_path.name

        try:
            with mrcfile.open(mrc_path) as mrc:
                img_data = mrc.data.astype(np.float64)

            print(f"Successfully loaded {mrc_filename} with shape {img_data.shape}.")
            micrographs.append((img_data, mrc_filename))

        except Exception as e:
            print(f"WARNING: Could not load or read MRC file {mrc_path}. Skipping. Error: {e}")
            continue

    return micrographs

def load_particle_coor(cfg: SNR_config, coor_path: Path, img_height: int) -> np.ndarray:
    """
    Loads particle coordinates for a micrograph from a CSV file.
    """
    print(f"Loading particle coordinates from {coor_path}...")
    try:
        # Read the CSV file
        with open(coor_path, 'r', newline='') as csvfile:
            reader = csv.reader(csvfile)
            header = next(reader)

            # Find the indices of the required columns
            try:
                x_idx = header.index('X-Coordinate')
                y_idx = header.index('Y-Coordinate')
            except ValueError:
                raise ValueError("CSV file header must contain 'X-Coordinate' and 'Y-Coordinate' columns.")

            coords = []
            for row in reader:
                try:
                    # Append coordinates, ensuring they are cast to floats
                    x = float(row[x_idx])
                    y = float(row[y_idx])
                    coords.append((x, y))
                except (ValueError, IndexError):
                    # Skip rows that don't have valid coordinate data
                    print(f"Warning: Skipping invalid row in {coor_path}: {row}")
                    continue

        if not coords:
            print("WARNING: Coordinate file was read but contained no valid particle data.")
            return np.array([])

        coords = np.array(coords, dtype=np.float64)

        if cfg.flip_y:
            # Flip the y-coordinates using the provided height
            coords[:, 1] = img_height - 1 - coords[:, 1]
            print(f"Y-axis for coordinates has been flipped using image height {img_height}.")

        return coords

    except FileNotFoundError:
        print(f"ERROR: Coordinate file not found at {coor_path}. Returning empty array.")
        return np.array([])
    except Exception as e:
        print(f"ERROR: Failed to load coordinates from {coor_path}. Error: {e}")
        return np.array([])

def load_mask(cfg: SNR_config, mask_path: Path) -> np.ndarray:
    """
    Loads a PNG mask file. Particles are set to 1.0, background to 0.0,
    """
    print(f"Loading mask from {mask_path}...")
    try:
        with Image.open(mask_path) as img:
            mask_data = np.array(img.convert('L'), dtype=np.float32)

        mask_data[mask_data > 0.1] = 1.0
        mask_data[mask_data <= 0.1] = 0.0
        return mask_data

    except FileNotFoundError:
        print(f"ERROR: Mask file not found at {mask_path}. Returning empty array.")
        return np.array([])
    except Exception as e:
        print(f"ERROR: Failed to load mask from {mask_path}. Error: {e}")
        return np.array([])

def SNR_csv_write_micrograph_data(cfg: SNR_config, mrc_filename: str, method: str,
                                  mean_micrograph_snr: float, total_pairs_N: int,
                                  file_name: str = "micrograph_snr_results.csv"):
    """
    Writes the single mean SNR result for a micrograph to a CSV file in the OUTPUT_DIR.
    """
    output_path = cfg.OUTPUT_DIR / file_name
    file_exists = output_path.exists()

    try:
        with open(output_path, mode='a', newline='') as file:
            writer = csv.writer(file)

            if not file_exists:
                writer.writerow([
                    "Micrograph_Filename",
                    "Method",
                    "Total_Pairs_N",
                    "Mean_SNR_Micrograph"
                ])

            writer.writerow([
                mrc_filename,
                method,
                total_pairs_N,
                f"{mean_micrograph_snr:.6f}"
            ])

        print(f"Result for {mrc_filename} recorded in {output_path.name}.")
    except Exception as e:
        print(f"ERROR: Failed to write to CSV file {output_path}. Error: {e}")


# --- Core SNR Calculation Functions ---

def _extract_block(img_data: np.ndarray, x: int, y: int, size: int) -> np.ndarray:
    """
    Extracts a square block of 'size' x 'size' pixels centered at (x, y).
    """
    h, w = img_data.shape

    if size % 2 == 0: # Even
        half_size = size // 2

        x_start = max(0, x - (half_size - 1))
        x_end = min(w, x + half_size)
        y_start = max(0, y - half_size)
        y_end = min(h, y + (half_size - 1))

    else:   # Odd
        size_adj = size - 1
        half_size = size_adj // 2

        x_start = max(0, x - half_size)
        x_end = min(w, x + half_size)
        y_start = max(0, y - half_size)
        y_end = min(h, y + half_size)

    block = img_data[y_start:y_end, x_start:x_end]

    return block

def SNR_pre_computation(particle_region: np.ndarray, background_region: np.ndarray) -> float:
    """
    SNR: log10( Var(Particle - Mean(Background)) / Var(Background) )
    """
    #  background statistics
    v_b = np.var(background_region)
    mu_b = np.mean(background_region)

    # Prevent division by zero
    if v_b <= 1e-12:
        return 0.0

    s = particle_region - mu_b
    v_s = np.var(s)

    snr_i = math.log10(v_s / v_b)
    return snr_i


# --- Helper for Distance Transform ---

def _calculate_distance_map(pf_binary_mask: np.ndarray) -> np.ndarray:
    """
    Calculates the minimum Euclidean distance from every background pixel (0)
    to the nearest particle pixel (1) in the mask.
    """
    # Find all coordinates marked as particles B( x, y ) = 1
    particle_pixels_y, particle_pixels_x = np.where(pf_binary_mask == 1)
    particle_pixel_coords = np.vstack((particle_pixels_x, particle_pixels_y)).T

    if particle_pixel_coords.size == 0:
        return np.full(pf_binary_mask.shape, np.inf, dtype=np.float32)

    # 1. Create a KDTree of all particle pixels
    particle_tree = KDTree(particle_pixel_coords)

    # 2. Create coordinates for all pixels in the image
    h, w = pf_binary_mask.shape
    y_coords, x_coords = np.mgrid[0:h, 0:w]
    all_pixel_coords = np.vstack((x_coords.ravel(), y_coords.ravel())).T

    # 3. Query the KDTree for the distance from every pixel to the nearest particle pixel
    distances_flat, _ = particle_tree.query(all_pixel_coords, k=1)

    # Reshape back to the image dimensions ( the Distance Map D(x,y) )
    distance_map = distances_flat.reshape(h, w)

    return distance_map

def plot_selected_regions(
    cfg: SNR_config,
    img_data: np.ndarray,
    img_filename: str,
    particle_centers: list,
    background_centers: list
):
    """
    Generates and saves a plot showing the selected particle and background blocks.

    - img_data: The original micrograph image data.
    - img_filename: The name of the micrograph file.
    - particle_centers: A list of (x, y) coordinates for particle blocks.
    - background_centers: A list of (x, y) coordinates for background blocks.
    """
    output_path = cfg.OUTPUT_DIR / "plots_selected_for_SNR" / f"{Path(img_filename).stem}_blocks.png"

    plt.figure(figsize=(12, 12))
    plt.imshow(img_data, cmap='gray')
    ax = plt.gca()

    block_size = cfg.block_size
    half_block = block_size / 2

    # Plotting particle blocks (red)
    for x, y in particle_centers:
        rect = Rectangle(
            (x - half_block, y - half_block),
            block_size,
            block_size,
            linewidth=1.5,
            edgecolor='r',
            facecolor='none',
            alpha=0.8
        )
        ax.add_patch(rect)

    # Plotting background blocks (blue)
    for x, y in background_centers:
        rect = Rectangle(
            (x - half_block, y - half_block),
            block_size,
            block_size,
            linewidth=1.5,
            edgecolor='b',
            facecolor='none',
            alpha=0.8
        )
        ax.add_patch(rect)

    plt.title(f"Selected Blocks for {img_filename}", fontsize=16)
    plt.xlabel("X-coordinate")
    plt.ylabel("Y-coordinate")
    #plt.axis('off')
    plt.tight_layout()
    plt.savefig(output_path, dpi=300)
    plt.show()
    #plt.close()
    print(f"  Plot of selected regions saved to {output_path}")

# --- Version 1: Block-Based (Smart Distance Sampling) Logic ---

def _remove_particle_blocks(cfg: SNR_config, np_img_data: np.ndarray, particles_coor: np.ndarray) -> Tuple[np.ndarray, np.ndarray]:
    """
    Creates a binary mask (1=particle, 0=background) for the Distance Transform,
    and the content mask (NaNs) for content exclusion.
    """
    h, w = np_img_data.shape
    # Initialize mask to 0 (all background)
    pf_binary_mask = np.zeros((h, w), dtype=np.int8)
    pf_size = cfg.PF_block_size

    if pf_size % 2 == 0:  # Even
        pf_half = pf_size // 2
        pf_half_adj = pf_half - 1

    else:  # Odd
        pf_size_adj = pf_size - 1
        pf_half = pf_size_adj // 2
        pf_half_adj = pf_half

    masked_img_content = np_img_data.copy()
    mask_value = np.nan

    for x, y in particles_coor:
        x, y = int(x), int(y)

        # bounding box
        x_start = max(0, x - pf_half_adj)
        x_end = min(w, x + pf_half)
        y_start = max(0, y - pf_half)
        y_end = min(h, y + pf_half_adj)

        # 1. binary mask update
        pf_binary_mask[y_start:y_end, x_start:x_end] = 1

        # 2. Update the content mask (NaNs for background extraction)
        masked_img_content[y_start:y_end, x_start:x_end] = mask_value

    # Return both the content mask and the binary mask for distance transform
    return masked_img_content, pf_binary_mask

def _pick_background_coords(cfg: SNR_config, masked_img_content: np.ndarray, pf_binary_mask: np.ndarray) -> List[Tuple[int, int]]:
    """
    Background selection blocks.
    """
    print("  Using Smart (Distance-based) Background Selection (Distance Map)...")
    h, w = masked_img_content.shape

    # 1. Calculate the Distance Map
    distance_map = _calculate_distance_map(pf_binary_mask)

    # 2. Define the valid sampling boundary based on block size
    size = cfg.block_size
    if size % 2 == 0:
        half_floor = size // 2 - 1
    else:
        half_floor = size // 2

    min_x = cfg.border_size + half_floor
    max_x = w - cfg.border_size - half_floor
    min_y = cfg.border_size + half_floor
    max_y = h - cfg.border_size - half_floor

    if max_x <= min_x or max_y <= min_y:
        print(f"  Warning: Image size ({w}x{h}) is too small for boundary constraints.")
        return []

    # 3. Filter potential centers based on Distance Map, Content Mask, and Mask Purity

    # Threshold: We require D(x,y) > Block Size Buffer (cfg.block_size * EDGE_CLEARANCE_FACTOR)
    min_required_clearance = cfg.block_size * EDGE_CLEARANCE_FACTOR

    # Mask Purity Threshold: Max allowed particle contamination in the background block
    MAX_MASK_CONTAMINATION = 0.03

    # 4. Generate candidate coordinates that satisfy all conditions
    clean_centers = []
    SAMPLE_STEP = 8

    for y in range(min_y, max_y, SAMPLE_STEP):
        for x in range(min_x, max_x, SAMPLE_STEP):
            # Check B: Distance Map value must exceed the clearance threshold
            if (distance_map[y, x] > min_required_clearance and
                distance_map[y, x] <= cfg.max_background_particles_distance):

                # Check C (NEW): Validate mask purity within the proposed block region
                # The _extract_block call handles boundary checks internally by returning zeros if out of bounds.
                mask_block = _extract_block(pf_binary_mask, x, y, cfg.block_size)

                # Calculate mean: Since the mask is 0/1, mean gives the fraction of particle pixels
                if np.mean(mask_block) < MAX_MASK_CONTAMINATION:
                    clean_centers.append((x, y))

    print(f"  Found {len(clean_centers)} centers with distance map clearance and mask purity (<{MAX_MASK_CONTAMINATION}).")

    if not clean_centers:
        print("  ERROR: No centers met the edge-clearance and purity thresholds. Returning empty list.")
        return []

    # 5. Filter for Non-Overlap: Apply the non-overlap constraint on the clean set
    final_coords = []
    rng = np.random.default_rng(cfg.random_state)
    rng.shuffle(clean_centers)

    min_center_dist = cfg.block_size + cfg.min_block_distance

    for x, y in clean_centers:
        if len(final_coords) >= cfg.blocks_num:
            break

        is_too_close = False
        for px, py in final_coords:
            dist_x = abs(x - px)
            dist_y = abs(y - py)

            # Check if the blocks are too close based on their center distance
            if dist_x < min_center_dist and dist_y < min_center_dist:
                is_too_close = True
                break

        if not is_too_close:
            final_coords.append((x, y))

    print(f"  Selected {len(final_coords)} distance-map, non-overlapping background blocks.")

    return final_coords


def SNR_within_blocks_block_based(cfg: SNR_config, np_img_data: np.ndarray, particles_coor: np.ndarray, img_filename: str) -> Tuple[float, int]:
    """
    Computes DNR by block
    """
    num_to_sample = cfg.blocks_num

    if len(particles_coor) < cfg.blocks_num: # Still use blocks_num for initial warning
        print(f"  WARNING: Not enough particles ({len(particles_coor)}) to select {cfg.blocks_num} pairs.")


    # 1. Randomly select particle centers (up to the maximum needed)
    rng = np.random.default_rng()
    num_particles_available = len(particles_coor)

    current_num_sets = min(num_to_sample, num_particles_available)

    if current_num_sets == 0:
        return 0.0, 0

    # Select required number of unique particle indices
    indices = rng.choice(num_particles_available, size=current_num_sets, replace=False)
    particle_centers = particles_coor[indices]

    # 2. Prepare image for background picking: get content mask and binary mask
    masked_img_content, pf_binary_mask = _remove_particle_blocks(cfg, np_img_data, particles_coor)

    # 3. Pick background block coordinates using the Distance Map and Purity Check
    background_centers_all = _pick_background_coords(cfg, masked_img_content, pf_binary_mask)

    # Match the number of available particles with available background centers
    current_num_sets = min(current_num_sets, len(background_centers_all))

    if current_num_sets == 0:
        print("  ERROR: Failed to select any valid pairs. Returning mean=0.0, N=0.")
        return 0.0, 0

    # Randomly select the required number of background blocks
    background_centers = random.sample(background_centers_all, current_num_sets)
    particle_centers = particle_centers[:current_num_sets] # Use only the corresponding particle centers

    if cfg.do_plots:
        plot_selected_regions(cfg, np_img_data, img_filename, particle_centers, background_centers)

    # 4. Extract blocks and compute SNR
    snr_list = []
    for i in range(current_num_sets):
        px, py = particle_centers[i]
        # Use _extract_block which now handles size precisely
        particle_region = _extract_block(np_img_data, int(px), int(py), cfg.block_size)

        bx, by = background_centers[i]
        # Use _extract_block which now handles size precisely
        background_region = _extract_block(np_img_data, bx, by, cfg.block_size)

        # Only compute if blocks are valid (not zero arrays)
        if particle_region.size > 0 and not np.all(particle_region == 0) and \
           background_region.size > 0 and not np.all(background_region == 0):
            snr_i = SNR_pre_computation(particle_region, background_region)
            snr_list.append(snr_i)

    N = len(snr_list)
    snr_val = 10 * np.mean(snr_list) if N > 0 else 0.0

    # NEW: Append the micrograph mean SNR to the config's SNR_list
    if N > 0:
        cfg.SNR_list.append(snr_val)

    print(f"  Successfully computed mean SNR of {N} individual SNR ratios: {snr_val:.4f}")
    return snr_val, N

# --- Version 2: Mask-Based Logic (Now uses smart non-overlap filtering) ---

def SNR_within_blocks_mask_based(cfg: SNR_config, np_img_data: np.ndarray, mask_data: np.ndarray, img_filename: str) -> Tuple[float, int]:
    """
    Compute SNR by Given Mask
    """
    print(f"  Starting Mask-Based selection (sampling up to {cfg.blocks_num} pairs)...")
    if mask_data.size == 0 or np_img_data.shape != mask_data.shape:
        print("  ERROR: Mask data is empty or dimensions mismatch. Returning mean=0.0, N=0.")
        return 0.0, 0

    h, w = np_img_data.shape
    block_size = cfg.block_size

    # 1. Define the valid sampling range
    if block_size % 2 == 0:
        half_floor = block_size // 2 - 1
    else:
        half_floor = block_size // 2

    valid_x_range = range(cfg.border_size + half_floor, w - cfg.border_size - half_floor)
    valid_y_range = range(cfg.border_size + half_floor, h - cfg.border_size - half_floor)

    particle_centers_all = []
    background_centers_all = []

    # Iterate over all possible block centers
    for x in valid_x_range:
        for y in valid_y_range:
            try:
                mask_block = _extract_block(mask_data, x, y, block_size)
            except ValueError:
                continue

            # Check for pure particle block (all pixels must be 1.0)
            if np.all(np.isclose(mask_block, 1.0)):
                particle_centers_all.append((x, y))

            # Check for pure background block (all pixels must be 0.0)
            elif np.all(np.isclose(mask_block, 0.0)):
                background_centers_all.append((x, y))


    # 2. Filter for Non-Overlap on both particle and background centers
    rng = np.random.default_rng(cfg.random_state)
    min_center_dist = cfg.block_size + cfg.min_block_distance

    def filter_non_overlap(centers: List[Tuple[int, int]]) -> List[Tuple[int, int]]:
        """Filters a list of centers to ensure no two selected centers are too close."""
        if not centers: return []

        # Shuffle for random sampling priority before filtering
        centers = centers.copy()
        rng.shuffle(centers)

        filtered_coords = []
        for x, y in centers:
            if len(filtered_coords) >= cfg.blocks_num:
                break

            is_too_close = False
            for px, py in filtered_coords:
                dist_x = abs(x - px)
                dist_y = abs(y - py)

                if dist_x < min_center_dist and dist_y < min_center_dist:
                    is_too_close = True
                    break

            if not is_too_close:
                filtered_coords.append((x, y))
        return filtered_coords

    # Apply non-overlap filter
    particle_centers_filtered = filter_non_overlap(particle_centers_all)
    background_centers_filtered = filter_non_overlap(background_centers_all)

    print(f"  Mask Filtered: {len(particle_centers_filtered)} non-overlapping particle centers.")
    print(f"  Mask Filtered: {len(background_centers_filtered)} non-overlapping background centers.")


    # 3. Determine the final number of pairs we can generate
    current_num_sets = min(len(particle_centers_filtered), len(background_centers_filtered))

    if current_num_sets == 0:
        print("  ERROR: Mask-based picking failed after filtering. Returning mean=0.0, N=0.")
        return 0.0, 0

    # 4. Final random sampling (since they are already non-overlapping, this just selects 'current_num_sets')
    particle_centers_sampled = random.sample(particle_centers_filtered, current_num_sets)
    background_centers_sampled = random.sample(background_centers_filtered, current_num_sets)

    if cfg.do_plots:
        plot_selected_regions(cfg, np_img_data, img_filename, particle_centers_sampled, background_centers_sampled)

    snr_list = []

    for i in range(current_num_sets):
        px, py = particle_centers_sampled[i]
        particle_region = _extract_block(np_img_data, px, py, block_size)

        bx, by = background_centers_sampled[i]
        background_region = _extract_block(np_img_data, bx, by, block_size)

        # Only compute if blocks are valid (not zero arrays)
        if particle_region.size > 0 and not np.all(particle_region == 0) and \
           background_region.size > 0 and not np.all(background_region == 0):
            snr_i = SNR_pre_computation(particle_region, background_region)
            snr_list.append(snr_i)

    N = len(snr_list)
    snr_val = 10 * np.mean(snr_list) if N > 0 else 0.0

    if N > 0:
        cfg.SNR_list.append(snr_val)

    print(f"  Successfully computed mean SNR of {N} individual SNR ratios: {snr_val:.4f}")
    return snr_val, N


# --- Main Class and Orchestration ---

class SNRCalculator:
    """
    SNR Class for user to call.
    """
    def __init__(self, config: SNR_config = SNR_config):
        config.mrc_list = []
        config.mrc_list_picked = []
        if config.blocks_num < 1:
            # Changed minimum sets to 1 for single mean calculation
            raise ValueError(f"blocks_num must be at least 1 for mean calculation. Got {config.blocks_num}")

    def calculate_final_stats(self, config: SNR_config = SNR_config):
        """
        Computes, prints, and records final statistics on the collected
        micrograph mean SNRs.
        """
        samples = np.array(config.SNR_list)
        N = len(samples)

        if N < 2:
            output = f"\n--- Final Statistics (Skipped) ---\n" \
                     f"Need at least 2 micrograph mean SNRs to compute CI/Box Plot data. Found N={N}.\n" \
                     f"---------------------------------"
            print(output)
            return

        # Prepare the output string to be both printed and written to file
        output_lines = []

        output_lines.append("\n" + "=" * 40)
        output_lines.append("Final Statistical Summary (Across all Micrographs)")
        output_lines.append("=" * 40)

        # 1. 95% Confidence Interval (t-distribution)
        mean_all = np.mean(samples)
        std_err = stats.sem(samples)
        degrees_freedom = N - 1

        if std_err == 0:
            ci_lower = ci_upper = mean_all
        else:
            ci_margin = std_err * t.ppf(0.975, degrees_freedom)
            ci_lower = mean_all - ci_margin
            ci_upper = mean_all + ci_margin

        output_lines.append(f"Total Micrographs (N): {N}")
        output_lines.append(f"Overall Mean SNR: {mean_all:.6f}")
        output_lines.append(f"95% Confidence Interval: ({ci_lower:.4f}, {ci_upper:.4f})")

        # 2. Visualization Data (Box Plot / Density)
        output_lines.append("\n[Visualization Data for Box/Density Plots]")
        output_lines.append("Sample distribution statistics (using Micrograph Mean SNRs as samples):")
        output_lines.append(f"  Median: {np.median(samples):.4f}")
        output_lines.append(f"  25th Percentile (Q1): {np.percentile(samples, 25):.4f}")
        output_lines.append(f"  75th Percentile (Q3): {np.percentile(samples, 75):.4f}")
        output_lines.append("  Full Sample List (for plotting density/box plot):")

        # Print a snippet or full list based on size for reference
        if N > 10:
            sample_str = f"    {samples[:5].tolist()} ... (showing first 5)"
        else:
            sample_str = f"    {samples.tolist()}"
        output_lines.append(sample_str)
        output_lines.append("---------------------------------")
        output_lines.append("\n" + "=" * 60)
        output_lines.append(f"Overall Mean Micrograph SNR: {mean_all:.6f}")
        output_lines.append("=" * 60)

        # Join the lines into a single string for printing and saving
        final_output_str = "\n".join(output_lines)
        print(final_output_str)

        # Write the same output to the file
        output_path_txt = config.OUTPUT_DIR / "SNR_rst.txt"
        try:
            with open(output_path_txt, 'w') as f:
                f.write(final_output_str)
            print(f"Final summary written to {output_path_txt}")
        except Exception as e:
            print(f"ERROR: Failed to write overall summary to {output_path_txt}. Error: {e}")


    def mean_SNR_within_micrographs(self, data_name: str, method: str = 'block_based', config: SNR_config = SNR_config) -> float:
        """
        Loops through all micrographs, gets SNR.
        """
        # Pass config to load_mrcfile
        if config.mrc_list != []:
            data_list = config.mrc_list
        else:
            data_list = load_mrcfile(config)

        temp_img_da, temp_img_name = data_list[0]
        h, w = temp_img_da.shape

        n_total_micrographs = len(data_list)
        if n_total_micrographs == 0:
            print("No micrograph data loaded. Exiting.")
            return 0.0

        # pick micrograph to use
        if config.num_micrograph_for_SNR is not None and config.num_micrograph_for_SNR < n_total_micrographs:
            num_to_sample = config.num_micrograph_for_SNR
        # Use a reproducible seed for random sampling
            rng = random.Random(config.random_state)
            sampled_micrographs = rng.sample(data_list, num_to_sample)
            print(f"Randomly selected {num_to_sample} micrographs for SNR calculation.")
        else:
            num_to_sample = n_total_micrographs
            sampled_micrographs = data_list
            print(f"Using all {n_total_micrographs} micrographs for SNR calculation.")

        print(f"\nProcessing {num_to_sample} micrographs using {method} method...")

        config.SNR_list = []
        config.mrc_list_picked = sampled_micrographs

        for img_data, img_filename in sampled_micrographs:
            print(f"\n--- Micrograph: {img_filename} ---")

            file_stem = Path(img_filename).stem
            micrograph_mean_snr = 0.0
            N = 0

            try:
                if method == 'block_based':
                    # The coordinate file is now a CSV with the same stem
                    coor_path = config.PARTICLE_COOR_DIR / f"{file_stem}.csv"
                    # Pass config
                    particles_coor = load_particle_coor(config, coor_path, h)
                    micrograph_mean_snr, N = SNR_within_blocks_block_based(config, img_data, particles_coor, img_filename)

                elif method == 'mask_based':
                    # Use config
                    mask_path = config.MASK_DIR / f"{file_stem}.png"
                    # Pass config
                    mask_data = load_mask(config, mask_path)  # I should change this part and make the mask wil also use the load particle coordinate
                    micrograph_mean_snr, N = SNR_within_blocks_mask_based(config, img_data, mask_data, img_filename)

                else:
                    print(f"ERROR: Unknown method '{method}'. Skipping.")
                    continue
            except Exception as e:
                print(f"CRITICAL ERROR during SNR calculation for {img_filename}: {e}. Skipping micrograph.")
                continue

            if N == 0:
                print(f"WARNING: No SNR values computed for this micrograph. Skipping further analysis.")
                continue

            # The mean is appended to cfg.SNR_list inside the calculation functions.

            # 2. Per-Micrograph CSV Writing (Simplified)
            SNR_csv_write_micrograph_data(
                cfg=config,
                mrc_filename=img_filename,
                method=method,
                mean_micrograph_snr=micrograph_mean_snr,
                total_pairs_N=N,
            )


        self.calculate_final_stats()


        # 4. Final Output (Mean of all micrograph means, stored in SNR_rst.txt)
        if not config.SNR_list:
             mean_mSNR = 0.0
        else:
             mean_mSNR = np.mean(config.SNR_list)

        return mean_mSNR
```
