```python
def preprocess_and_crop(micrograph, crop_size=3840):
    processed_micrograph = simple_micrograph_preprocessing(micrograph)
    if crop_size:
        mic_width, mic_height = processed_micrograph.shape[1], processed_micrograph.shape[0]
        start_x, start_y = (mic_width - crop_size) // 2, (mic_height - crop_size) // 2
        end_x, end_y = start_x + crop_size, start_y + crop_size
        return processed_micrograph[start_y:end_y, start_x:end_x]
    else:
        return processed_micrograph

def plot_micrograph_and_labels(ax, micrograph, labels, coords):
    ax.imshow(micrograph, cmap='gray')
    ax.imshow(labels, cmap='gray', alpha=0.5)
    for x, y in coords:
        corrected_x, corrected_y = x, y
        circle = matplotlib.patches.Circle((corrected_x, corrected_y), radius=RADIUS, fill=False, color='r')
        ax.add_patch(circle)
```

## DT with circularity, Sphericity, Slenderness at scalar for the mean of front 50% 0.65

```python
# @title particle finding function of watershed
import cv2
from scipy import ndimage as ndi
from scipy.spatial.distance import mahalanobis
from skimage.feature import peak_local_max
from skimage.segmentation import watershed
from skimage.measure import regionprops

""" EXAMPLE OF ndimage.distance_transform_edt()
a = np.array(([0,1,1,1,1],
              [0,0,1,1,1],
              [0,1,1,1,1],
              [0,1,1,1,0],
              [0,1,1,0,0]))
ndimage.distance_transform_edt(a)
array([[ 0.    ,  1.    ,  1.4142,  2.2361,  3.    ],
       [ 0.    ,  0.    ,  1.    ,  2.    ,  2.    ],
       [ 0.    ,  1.    ,  1.4142,  1.4142,  1.    ],
       [ 0.    ,  1.    ,  1.4142,  1.    ,  0.    ],
       [ 0.    ,  1.    ,  1.    ,  0.    ,  0.    ]])
"""

def calculate_shape_metrics_extended(region):
    """Computes an expanded 5D feature set for multivariate analysis."""
    area = region.area
    perimeter = region.perimeter
    if perimeter == 0: return [0] * 5

    # 1. Circularity
    circularity = (4 * np.pi * area) / (perimeter ** 2)

    # 2. Sphericity
    d_eq = np.sqrt(4 * area / np.pi)
    sphericity = d_eq / region.major_axis_length if region.major_axis_length > 0 else 0

    # 3. Solidity (Area / Convex Area)
    solidity = region.solidity

    # 4. Convexity (Convex Perimeter / Perimeter) - measures boundary smoothness
    extent = region.extent

    return [sphericity, circularity, solidity, extent]


def detect_particles_dt(prob_map_input, particle_radius=64,
                        peak_threshold_ratio=0.4, min_dist_ratio=0.8,
                        z_thresh=3.0, show_plots=True):
    # 1-4. Preprocessing & Watershed
    prob_map = prob_map_input.astype(np.float32)
    height, width = prob_map.shape[:2]
    if prob_map.max() > 1.0: prob_map /= 255.0
    binary_mask = (prob_map > 0.5).astype(np.uint8)
    if np.sum(binary_mask) == 0: binary_mask = (prob_map > 0.05).astype(np.uint8)
    if np.sum(binary_mask) == 0: return []

    distance = ndi.distance_transform_edt(binary_mask)
    local_max_coords = peak_local_max(distance, min_distance=int(particle_radius * min_dist_ratio),
                                      labels=binary_mask, threshold_abs=particle_radius * peak_threshold_ratio)
    markers = np.zeros(distance.shape, dtype=int)
    for i, (r, c) in enumerate(local_max_coords): markers[r, c] = i + 1
    labels = watershed(-distance, markers, mask=binary_mask)

    # 5. Candidate Extraction (4D Features)
    expected_area = np.pi * (particle_radius ** 2)
    min_area, max_area = expected_area * 0.3, expected_area * 2.0
    regions = regionprops(labels)

    candidates, feature_list = [], []
    for region in regions:
        if min_area <= region.area <= max_area:
            features = calculate_shape_metrics_extended(region) # Returns 4 metrics
            candidates.append(region)
            feature_list.append(features)

    if not feature_list: return []

    # 6. Mahalanobis Calculation
    X = np.array(feature_list)
    median_vec = np.median(X, axis=0)
    mad_vec = np.median(np.abs(X - median_vec), axis=0)
    std_proxy = mad_vec / 0.6745
    precision_matrix = np.diag(1.0 / (std_proxy**2 + 1e-6))

    dists = np.array([mahalanobis(f, median_vec, precision_matrix) for f in X])

    # 7. changed) Regularization & Range Strategy
    # Normalize distances to [0, 1]
    d_min, d_max = dists.min(), dists.max()
    d_norm = (dists - d_min) / (d_max - d_min + 1e-9) if d_max > d_min else np.zeros_like(dists)

    def get_mad_range(data, z):
        med = np.median(data)
        mad = np.median(np.abs(data - med))
        if mad == 0: return 0.0, 1.0
        # Range Strategy using adjusted Z-score
        low = med - (z * mad / 0.6745)
        high = med + (z * mad / 0.6745)
        return max(0.0, low), min(1.0, high)

    # Calculate adaptive range for the normalized distances
    t_min, t_max = get_mad_range(d_norm, z_thresh)

    if show_plots:
        plt.figure(figsize=(8, 5))
        plt.hist(d_norm, bins=30, color='teal', alpha=0.6)
        plt.axvline(t_min, color='red', linestyle='--', label='Range Min')
        plt.axvline(t_max, color='blue', linestyle='--', label='Range Max')
        plt.title(f"Normalized Mahalanobis Range (N={len(candidates)})")
        plt.legend()
        plt.show()

    # 8. Final Selection
    valid_particles = []
    boundary_margin = 40
    for i, val in enumerate(d_norm):
        # Apply the adaptive range filter
        if t_min <= val <= t_max:
            region = candidates[i]
            cy, cx = region.centroid
            if (boundary_margin < cx < width - boundary_margin and
                boundary_margin < cy < height - boundary_margin):
                score = prob_map[int(cy), int(cx)]
                valid_particles.append([int(cx), int(cy), float(score)])

    return valid_particles
```

```python
def detect_particles_dt_error_analysis(prob_map_input, particle_radius=64,
                                       peak_threshold_ratio=0.4, min_dist_ratio=0.8,
                                       z_thresh=3.0):
    # --- 1-4. Preprocessing & Watershed (Standard) ---
    prob_map = prob_map_input.astype(np.float32)
    height, width = prob_map.shape[:2]
    if prob_map.max() > 1.0: prob_map /= 255.0
    binary_mask = (prob_map > 0.5).astype(np.uint8)
    if np.sum(binary_mask) == 0: binary_mask = (prob_map > 0.05).astype(np.uint8)
    if np.sum(binary_mask) == 0:
        return {"valid": [], "area_filtered": [], "range_filtered": []}

    distance = ndi.distance_transform_edt(binary_mask)
    local_max_coords = peak_local_max(distance, min_distance=int(particle_radius * min_dist_ratio),
                                      labels=binary_mask, threshold_abs=particle_radius * peak_threshold_ratio)
    markers = np.zeros(distance.shape, dtype=int)
    for i, (r, c) in enumerate(local_max_coords): markers[r, c] = i + 1
    labels = watershed(-distance, markers, mask=binary_mask)

    # --- 5. Candidate Extraction & Area Filter ---
    expected_area = np.pi * (particle_radius ** 2)
    min_area, max_area = expected_area * 0.3, expected_area * 2.0
    regions = regionprops(labels)

    candidates, feature_list, area_filtered = [], [], []

    for region in regions:
        cy, cx = region.centroid
        # changed) logic to track area rejections
        if not (min_area <= region.area <= max_area):
            area_filtered.append([int(cx), int(cy)])
            continue

        features = calculate_shape_metrics_extended(region)
        candidates.append(region)
        feature_list.append(features)

    if not feature_list:
        return {"valid": [], "area_filtered": area_filtered, "range_filtered": []}

    # --- 6. Mahalanobis Calculation ---
    X = np.array(feature_list)
    median_vec = np.median(X, axis=0)
    mad_vec = np.median(np.abs(X - median_vec), axis=0)
    std_proxy = mad_vec / 0.6745
    precision_matrix = np.diag(1.0 / (std_proxy**2 + 1e-6))
    dists = np.array([mahalanobis(f, median_vec, precision_matrix) for f in X])

    # --- 7. Regularization & Range Calculation ---
    d_min, d_max = dists.min(), dists.max()
    d_norm = (dists - d_min) / (d_max - d_min + 1e-9) if d_max > d_min else np.zeros_like(dists)

    med_n = np.median(d_norm)
    mad_n = np.median(np.abs(d_norm - med_n))
    t_min = max(0.0, med_n - (z_thresh * mad_n / 0.6745))
    t_max = min(1.0, med_n + (z_thresh * mad_n / 0.6745))

    # --- 8. Final Categorization ---
    valid, range_filtered = [], []
    boundary_margin = 40

    for i, val in enumerate(d_norm):
        region = candidates[i]
        cy, cx = region.centroid

        # changed) multivariate range check
        if t_min <= val <= t_max:
            if (boundary_margin < cx < width - boundary_margin and
                boundary_margin < cy < height - boundary_margin):
                score = prob_map[int(cy), int(cx)]
                valid.append([int(cx), int(cy), float(score)])
            else:
                area_filtered.append([int(cx), int(cy)]) # Boundary rejections
        else:
            range_filtered.append([int(cx), int(cy)])

    return {
        "valid": valid,
        "area_filtered": area_filtered,
        "range_filtered": range_filtered
    }
```

```python
def plot_micrograph_error_analysis(ax, micrograph, labels, particle_dict):
    ax.imshow(micrograph, cmap='gray')
    ax.imshow(labels, cmap='gray', alpha=0.3)

    # Area Rejections (Limegreen)
    if particle_dict['area_filtered']:
        pts = np.array(particle_dict['area_filtered'])
        ax.scatter(pts[:, 0], pts[:, 1], marker='x', color='limegreen', s=100, label='Area Rej')

    # Range Rejections (Navy)
    if particle_dict['range_filtered']:
        pts = np.array(particle_dict['range_filtered'])
        ax.scatter(pts[:, 0], pts[:, 1], marker='x', color='navy', s=100, label='Shape Outlier')

    # Valid (Red)
    if particle_dict['valid']:
        pts = np.array(particle_dict['valid'])
        ax.scatter(pts[:, 0], pts[:, 1], marker='+', color='red', s=150, linewidths=1.5, label='Valid')

    ax.legend(loc='upper right', frameon=True)
```

```python
def detect_particles_dt_filtered(prob_map_input, particle_radius=64,
                                       peak_threshold_ratio=0.4, min_dist_ratio=0.8,
                                       z_thresh=3.0):
    # --- 1-4. Preprocessing & Watershed ---
    prob_map = prob_map_input.astype(np.float32)
    height, width = prob_map.shape[:2]
    if prob_map.max() > 1.0: prob_map /= 255.0
    binary_mask = (prob_map > 0.5).astype(np.uint8)
    if np.sum(binary_mask) == 0: binary_mask = (prob_map > 0.05).astype(np.uint8)
    if np.sum(binary_mask) == 0:
        return {"valid": [], "area_filtered": [], "shape_filtered": []}

    distance = ndi.distance_transform_edt(binary_mask)
    local_max_coords = peak_local_max(distance, min_distance=int(particle_radius * min_dist_ratio),
                                      labels=binary_mask, threshold_abs=particle_radius * peak_threshold_ratio)
    markers = np.zeros(distance.shape, dtype=int)
    for i, (r, c) in enumerate(local_max_coords): markers[r, c] = i + 1
    labels = watershed(-distance, markers, mask=binary_mask)

    # --- 5. Initial Area Filtering ---
    expected_area = np.pi * (particle_radius ** 2)
    min_area, max_area = expected_area * 0.3, expected_area * 2.0
    regions = regionprops(labels)

    candidates = []
    area_filtered = []
    feature_list = []

    for region in regions:
        cy, cx = region.centroid
        # changed) logic to separate area rejections
        if not (min_area <= region.area <= max_area):
            area_filtered.append([int(cx), int(cy)])
            continue

        features = calculate_shape_metrics_extended(region)
        candidates.append({'region': region, 'features': features})
        feature_list.append(features)

    if not candidates:
        return {"valid": [], "area_filtered": area_filtered, "shape_filtered": []}

    # --- 6. Multivariate Mahalanobis Calculation ---
    X = np.array(feature_list)
    median_vec = np.median(X, axis=0)
    mad_vec = np.median(np.abs(X - median_vec), axis=0)
    std_proxy = mad_vec / 0.6745
    precision_matrix = np.diag(1.0 / (std_proxy**2 + 1e-6))

    mahal_distances = np.array([mahalanobis(feat, median_vec, precision_matrix) for feat in X])

    # --- 7. Final Categorization ---
    valid = []
    shape_filtered = []

    for i, cand in enumerate(candidates):
        region = cand['region']
        cy, cx = region.centroid

        # changed) Multivariate Shape filter using Mahalanobis distance
        if mahal_distances[i] > z_thresh:
            shape_filtered.append([int(cx), int(cy)])
        else:
            score = prob_map[int(cy), int(cx)]
            valid.append([int(cx), int(cy), float(score)])

    return {
        "valid": valid,
        "area_filtered": area_filtered,
        "shape_filtered": shape_filtered
    }
```

```python
def detect_particles_dt_filtered(prob_map_input, particle_radius=64,
                                 peak_threshold_ratio=0.4, min_dist_ratio=0.8,
                                 z_thresh=3.0, show_plots=True):
    # 1. Normalization & Masking
    prob_map = prob_map_input.astype(np.float32)
    height, width = prob_map.shape[:2]
    if prob_map.max() > 1.0: prob_map /= 255.0

    binary_mask = (prob_map > 0.5).astype(np.uint8)
    if np.sum(binary_mask) == 0: binary_mask = (prob_map > 0.05).astype(np.uint8)
    if np.sum(binary_mask) == 0: return {"valid":[], "area_filtered":[], "shape_filtered":[]}

    # 2. Watershed Segmentation
    distance = ndi.distance_transform_edt(binary_mask)
    min_dist = int(particle_radius * min_dist_ratio)
    min_peak_height = particle_radius * peak_threshold_ratio

    local_max_coords = peak_local_max(distance, min_distance=min_dist,
                                      labels=binary_mask, threshold_abs=min_peak_height)
    markers = np.zeros(distance.shape, dtype=int)
    for i, (r, c) in enumerate(local_max_coords): markers[r, c] = i + 1
    labels = watershed(-distance, markers, mask=binary_mask)

    # 3. Candidate Extraction & Area Filter
    expected_area = np.pi * (particle_radius ** 2)
    min_area, max_area = expected_area * 0.3, expected_area * 2.0
    regions = regionprops(labels)

    candidates, area_filtered = [], []
    m_sp, m_cr, m_isd, m_sol, m_ext = [], [], [], [], []

    for region in regions:
        cy, cx = region.centroid
        if not (min_area <= region.area <= max_area):
            area_filtered.append([int(cx), int(cy)])
            continue

        # changed) Extracting the full 5D metric set
        f = calculate_shape_metrics_extended(region)
        candidates.append({'region': region, 'f': f})
        m_sp.append(f[0]); m_cr.append(f[1]); m_isd.append(f[2]); m_sol.append(f[3]); m_ext.append(f[4])

    if not candidates: return {"valid":[], "area_filtered":area_filtered, "shape_filtered":[]}

    # 4. changed) Adaptive Range Calculation (MAD Strategy)
    def get_mad_range(data, z):
        arr = np.array(data)
        med = np.median(arr)
        mad = np.median(np.abs(arr - med))
        if mad == 0: return np.min(arr), np.max(arr)
        low = med - (z * mad / 0.6745)
        high = med + (z * mad / 0.6745)
        return max(0.0, low), min(1.0, high)

    t_sp = get_mad_range(m_sp, z_thresh)
    t_cr = get_mad_range(m_cr, z_thresh)
    t_isd = get_mad_range(m_isd, z_thresh)
    t_sol = get_mad_range(m_sol, z_thresh)
    t_ext = get_mad_range(m_ext, z_thresh)

    # 5. changed) Multi-Metric Visualization (5 Histograms)
    if show_plots:
        fig, axes = plt.subplots(1, 5, figsize=(20, 4))
        metrics = [m_sp, m_cr, m_isd, m_sol, m_ext]
        titles = ['Sphericity', 'Circularity', 'Inv-Slenderness', 'Solidity', 'Extent']
        bounds = [t_sp, t_cr, t_isd, t_sol, t_ext]

        for i in range(5):
            axes[i].hist(metrics[i], bins=20, color='gray', alpha=0.5)
            axes[i].axvline(bounds[i][0], color='red', linestyle='--', label='Min')
            axes[i].axvline(bounds[i][1], color='blue', linestyle='--', label='Max')
            axes[i].set_title(titles[i])
            if i == 0: axes[i].legend()
        plt.suptitle(f"Univariate 5D Adaptive Thresholding (N={len(candidates)})")
        plt.tight_layout()
        plt.show()

    # 6. changed) Final Selection using AND Logic
    valid, shape_filtered = [], []
    boundary_margin = 40
    for cand in candidates:
        f = cand['f']
        # changed) AND Logic: Must pass all 5 metric bounds
        is_valid_shape = (t_sp[0] <= f[0] <= t_sp[1] and
                          t_cr[0] <= f[1] <= t_cr[1] and
                          t_isd[0] <= f[2] <= t_isd[1] and
                          t_sol[0] <= f[3] <= t_sol[1] and
                          t_ext[0] <= f[4] <= t_ext[1])

        region = cand['region']
        cy, cx = region.centroid

        if is_valid_shape:
            if (boundary_margin < cx < width - boundary_margin and
                boundary_margin < cy < height - boundary_margin):
                score = prob_map[int(cy), int(cx)]
                valid.append([int(cx), int(cy), float(score)])
            else:
                area_filtered.append([int(cx), int(cy)])
        else:
            shape_filtered.append([int(cx), int(cy)])

    # return {"valid": valid, "area_filtered": area_filtered, "shape_filtered": shape_filtered}
    return valid
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler, MinMaxScaler

def detect_particles_pca_range(prob_map_input, particle_radius=64,
                               peak_threshold_ratio=0.4, min_dist_ratio=0.8,
                               z_thresh=3.0, show_plots=True):
    # --- 1-4. Preprocessing & Watershed ---
    prob_map = prob_map_input.astype(np.float32)
    height, width = prob_map.shape[:2]
    if prob_map.max() > 1.0: prob_map /= 255.0
    binary_mask = (prob_map > 0.5).astype(np.uint8)
    if np.sum(binary_mask) == 0: binary_mask = (prob_map > 0.05).astype(np.uint8)
    if np.sum(binary_mask) == 0: return []

    distance = ndi.distance_transform_edt(binary_mask)
    local_max_coords = peak_local_max(distance, min_distance=int(particle_radius * min_dist_ratio),
                                      labels=binary_mask, threshold_abs=particle_radius * peak_threshold_ratio)
    markers = np.zeros(distance.shape, dtype=int)
    for i, (r, c) in enumerate(local_max_coords): markers[r, c] = i + 1
    labels = watershed(-distance, markers, mask=binary_mask)

    # --- 5. Candidate Feature Extraction (5D) ---
    regions = regionprops(labels)
    expected_area = np.pi * (particle_radius ** 2)
    min_area, max_area = expected_area * 0.3, expected_area * 2.0

    candidates, feature_list = [], []
    for region in regions:
        if min_area <= region.area <= max_area:
            features = calculate_shape_metrics_extended(region)
            candidates.append(region)
            feature_list.append(features)

    if not feature_list: return []

    # --- 6. PCA & Regularization [0, 1] ---
    # changed) Standardize then apply PCA (Dim=2)
    X = np.array(feature_list)
    X_scaled = StandardScaler().fit_transform(X)
    pca = PCA(n_components=2)
    X_pca = pca.fit_transform(X_scaled)

    # changed) Regularize PCA components to [0, 1] range
    scaler_01 = MinMaxScaler(feature_range=(0, 1))
    X_norm = scaler_01.fit_transform(X_pca)

    # --- 7. Adaptive Range Strategy (Using MAD-Adjusted Z-Score) ---
    def get_range_thresholds(data, z):
        # Calculate robust center (Median) and spread (MAD)
        med = np.median(data)
        mad = np.median(np.abs(data - med))
        if mad == 0: return 0.0, 1.0

        # changed) Range Strategy: Define adaptive bounds based on normalized spread
        # Adjusted Z-score using 0.6745 for consistency with standard deviation
        low = med - (z * mad / 0.6745)
        high = med + (z * mad / 0.6745)

        # Clamp to [0, 1] as requested
        return max(0.0, low), min(1.0, high)

    t_pc1_min, t_pc1_max = get_range_thresholds(X_norm[:, 0], z_thresh)
    t_pc2_min, t_pc2_max = get_range_thresholds(X_norm[:, 1], z_thresh)

    # Apply AND operator across components
    mask_pc1 = (X_norm[:, 0] >= t_pc1_min) & (X_norm[:, 0] <= t_pc1_max)
    mask_pc2 = (X_norm[:, 1] >= t_pc2_min) & (X_norm[:, 1] <= t_pc2_max)
    final_mask = mask_pc1 & mask_pc2

    # --- 8. Visualizing PCA Range Filter ---
    if show_plots:
        fig, ax = plt.subplots(figsize=(8, 6))
        ax.scatter(X_norm[final_mask, 0], X_norm[final_mask, 1], c='red', marker='+', label='Valid')
        ax.scatter(X_norm[~final_mask, 0], X_norm[~final_mask, 1], c='navy', marker='x', alpha=0.5, label='Filtered')

        # Draw the "Range Box"
        rect = plt.Rectangle((t_pc1_min, t_pc2_min), t_pc1_max - t_pc1_min, t_pc2_max - t_pc2_min,
                             fill=False, color='green', linestyle='--', label='Adaptive Range')
        ax.add_patch(rect)
        ax.set_title(f"PCA [0,1] Regularized Range Strategy (Z={z_thresh})")
        ax.set_xlabel("Normalized PC1"); ax.set_ylabel("Normalized PC2")
        ax.legend()
        plt.show()

    # --- 9. Final Selection ---
    valid_particles = []
    boundary_margin = 40
    for i, is_valid in enumerate(final_mask):
        if is_valid:
            region = candidates[i]
            cy, cx = region.centroid
            if (boundary_margin < cx < width - boundary_margin and
                boundary_margin < cy < height - boundary_margin):
                score = prob_map[int(cy), int(cx)]
                valid_particles.append([int(cx), int(cy), float(score)])

    return valid_particles
```

```python
from scipy import ndimage as ndi
from skimage.feature import peak_local_max
from skimage.segmentation import watershed
from skimage.measure import regionprops

def calculate_shape_metrics(region):
    """Computes Sphericity, Circularity, and Inverse Slenderness."""
    area = region.area
    perimeter = region.perimeter
    if perimeter == 0: return 0, 0, 0

    circularity = (4 * np.pi * area) / (perimeter ** 2)

    d_eq = np.sqrt(4 * area / np.pi)
    d_cir = region.major_axis_length
    sphericity = d_eq / d_cir if d_cir > 0 else 0

    return sphericity, circularity

def detect_particles_dt(prob_map_input, particle_radius=64,
                        peak_threshold_ratio=0.6, min_dist_ratio=0.4,
                        show_plots=True):
    # 1. Normalization
    prob_map = prob_map_input.astype(np.float32)
    height, width = prob_map.shape[:2]
    if prob_map.max() > 1.0: prob_map /= 255.0

    binary_mask = (prob_map > 0.5).astype(np.uint8)
    if np.sum(binary_mask) == 0: binary_mask = (prob_map > 0.05).astype(np.uint8)
    if np.sum(binary_mask) == 0: return []

    # 2. DT transform
    distance = ndi.distance_transform_edt(binary_mask)

    # 3. Markers Finding
    min_dist = int(particle_radius * min_dist_ratio)
    min_peak_height = particle_radius * peak_threshold_ratio

    local_max_coords = peak_local_max(
        distance,
        min_distance=min_dist,
        labels=binary_mask,
        threshold_abs=min_peak_height
    )

    markers = np.zeros(distance.shape, dtype=int)
    for i, (r, c) in enumerate(local_max_coords):
        markers[r, c] = i + 1

    # 4. Watershed
    labels = watershed(-distance, markers, mask=binary_mask)

    # 5. Extract Candidate Metrics
    expected_area = np.pi * (particle_radius ** 2)
    min_area, max_area = expected_area * 0.3, expected_area * 2.0

    regions = regionprops(labels)
    candidates = []
    m_sp, m_cr = [], []

    for region in regions:
        if min_area <= region.area <= max_area:
            sp, cr = calculate_shape_metrics(region)
            candidates.append({'region': region, 'metrics': (sp, cr)})
            m_sp.append(sp); m_cr.append(cr)

    if not candidates: return []

    # 6. Statistical Thresholding (Modified Z-Score / MAD)
    #  MAD statistics thresholding
    def get_mad_range(data, z_thresh=3.0):
        if not data: return 0.0, 1.0
        arr = np.array(data)
        med = np.median(arr)
        mad = np.median(np.abs(arr - med))
        if mad == 0: return np.min(arr), np.max(arr)

        # Calculate bounds based on 3 median absolute deviations
        low = med - (z_thresh * mad / 0.6745)
        high = med + (z_thresh * mad / 0.6745)

        return max(0.0, low), min(1.0, high)

    thresh_sp = get_mad_range(m_sp)
    thresh_cr = get_mad_range(m_cr)

    # 7. Auto-Visualization
    # changed) added automatic plotting of the current image's metric distributions
    if show_plots:
        fig, axes = plt.subplots(1, 2, figsize=(12, 4))
        metrics = [m_sp, m_cr]
        titles = ['Sphericity', 'Circularity']
        bounds = [thresh_sp, thresh_cr]

        for i in range(2):
            axes[i].hist(metrics[i], bins=20, color='gray', alpha=0.5)
            axes[i].axvline(bounds[i][0], color='red', linestyle='--', label='Min')
            axes[i].axvline(bounds[i][1], color='blue', linestyle='--', label='Max')
            axes[i].set_title(titles[i])
            if i == 0: axes[i].legend()
        plt.suptitle(f"Adaptive Thresholds for Current Image (N={len(candidates)})")
        plt.show()

    # 8. Final Selection
    valid_particles = []
    boundary_margin = 40
    for cand in candidates:
        sp, cr = cand['metrics']
        if (thresh_sp[0] <= sp <= thresh_sp[1] and
            thresh_cr[0] <= cr <= thresh_cr[1]):

            region = cand['region']
            cy, cx = region.centroid
            if (boundary_margin < cx < width - boundary_margin and
                boundary_margin < cy < height - boundary_margin):
                score = prob_map[int(cy), int(cx)]
                valid_particles.append([int(cx), int(cy), float(score)])

    return valid_particles
```

```python
def detect_particles_dt(prob_map_input, particle_radius=64,
                        peak_threshold_ratio=0.4, min_dist_ratio=0.8,
                        show_plots=False):
    # 1. Normalization
    prob_map = prob_map_input.astype(np.float32)
    height, width = prob_map.shape[:2]
    if prob_map.max() > 1.0: prob_map /= 255.0

    # 2. Binary Masking
    binary_mask = (prob_map > 0.5).astype(np.uint8)
    if np.sum(binary_mask) == 0: binary_mask = (prob_map > 0.05).astype(np.uint8)
    if np.sum(binary_mask) == 0: return []

    # 3. Distance Transform & Marker Finding
    distance = ndi.distance_transform_edt(binary_mask)
    min_dist = int(particle_radius * min_dist_ratio)
    min_peak_height = particle_radius * peak_threshold_ratio

    local_max_coords = peak_local_max(distance, min_distance=min_dist,
                                      labels=binary_mask, threshold_abs=min_peak_height)

    markers = np.zeros(distance.shape, dtype=int)
    for i, (r, c) in enumerate(local_max_coords): markers[r, c] = i + 1

    # 4. Watershed Segmentation
    labels = watershed(-distance, markers, mask=binary_mask)

    # 5. Candidate Extraction (Area Only)
    # changed) Removed all multivariate and univariate shape metrics.
    expected_area = np.pi * (particle_radius ** 2)
    min_area, max_area = expected_area * 0.3, expected_area * 2.0
    regions = regionprops(labels)

    valid_particles = []
    boundary_margin = 40

    for region in regions:
        if min_area <= region.area <= max_area:
            cy, cx = region.centroid

            # Boundary check to ensure particles are not cut off
            if (boundary_margin < cx < width - boundary_margin and
                boundary_margin < cy < height - boundary_margin):

                score = prob_map[int(cy), int(cx)]
                valid_particles.append([int(cx), int(cy), float(score)])

    return valid_particles
```

```python
import itertools

watershed_list_all = []
watershed_config = []

# Search Space
# threshold_levels_ratiros = [0.5, 0.6, 0.7]
threshold_levels_ratios = [0.4]

# min_dist_ratios = [0.8, 0.9, 1.0]
min_dist_ratios = [0.8]

print(f"Starting DT-Only Grid Search with Adaptive Shape Filtering...")

param_grid = list(itertools.product(
    threshold_levels_ratios, min_dist_ratios
))

for params in param_grid:
    thresh_r, dist_r = params

    print(f"Config: Thresh={thresh_r}, Dist={dist_r}")
    watershed_list = []

    for img in label_images:
        particles = detect_particles_dt(
            img,
            particle_radius=RADIUS,
            peak_threshold_ratio=thresh_r,
            min_dist_ratio=dist_r
        )
        watershed_list.append(particles)

    watershed_list_all.append(watershed_list)
    watershed_config.append((thresh_r, dist_r))

print(f" -> Finished generating {len(watershed_list_all)} candidates.")
```

```python
# @title DT-Only Grid Search: Evaluate & Score
import torch
import os
import json

# Metrics Helper
def evaluate_detection_robust(iou_matrices, iou_threshold=0.5):
    tp_total = 0; fp_total = 0; fn_total = 0
    for iou_matrix in iou_matrices:
        n_gt, n_pred = iou_matrix.shape
        if n_gt == 0 and n_pred == 0: continue
        if n_gt > 0 and n_pred == 0: fn_total += n_gt; continue
        if n_gt == 0 and n_pred > 0: fp_total += n_pred; continue

        max_vals, _ = torch.max(iou_matrix, dim=1)
        detected_mask = max_vals >= iou_threshold
        tp = torch.sum(detected_mask).item()
        tp_total += tp
        fn_total += (n_gt - tp)
        fp_total += (n_pred - tp)

    epsilon = 1e-6
    precision = tp_total / (tp_total + fp_total + epsilon)
    recall = tp_total / (tp_total + fn_total + epsilon)
    return precision, recall

# Main Eval Loop
width = RADIUS * 2
watershed_scores = []

print("Evaluating DT-Only Results...")

for i, ws_list in enumerate(watershed_list_all):
    cfg = watershed_config[i] # (thresh, dist)

    iou_matrices = []
    gt_full = []
    pred_full = []
    scores_list = []

    for idx, gt in enumerate(gts):
        gt_boxes = centers_to_boxes(np.array(gt), width, width)
        gt_full.append(gt_boxes)

        preds = ws_list[idx]
        if len(preds) > 0:
            coords = np.array([p[:2] for p in preds])
            sc = torch.tensor([p[2] for p in preds], dtype=torch.float32)
            pred_boxes = centers_to_boxes(coords, width, width)
        else:
            pred_boxes = torch.empty((0, 4))
            sc = torch.tensor([], dtype=torch.float32)

        pred_full.append(pred_boxes)
        scores_list.append(sc)

        iou_matrix = calculate_iou_torchvision(gt_boxes, pred_boxes)
        iou_matrices.append(iou_matrix)

    # 1. F-Beta
    precision, recall = evaluate_detection_robust(iou_matrices, iou_threshold=0.5)
    f_beta = f_beta_score(precision, recall, beta=1)

    # 2. mAP
    iou_thresholds = torch.arange(0.5, 1.0, 0.05)
    try:
        mAP_value = calculate_mAP_multiple_images(gt_full, pred_full, scores_list, iou_thresholds)
        map_val = mAP_value.item()
    except:
        map_val = 0.0

    print(f"Cfg: {cfg} | F1: {f_beta:.4f}, mAP: {map_val:.4f}")
    watershed_scores.append((f_beta, map_val))

# Save Best
if len(watershed_scores) > 0:
    best_idx = max(range(len(watershed_scores)), key=lambda i: watershed_scores[i][0])
    best_cfg = watershed_config[best_idx]

    best_res = {
        "method": "dt_only_no_ar",
        "thresh": best_cfg[0],
        "dist": best_cfg[1],
        "f_score": watershed_scores[best_idx][0],
        "map": watershed_scores[best_idx][1]
    }

    print("\n--- BEST CONFIGURATION (DT ONLY - NO AR) ---")
    print(best_res)

    # 1. Save to local content first (Current Directory)
    local_file = 'best_watershed_params.json'
    with open(local_file, 'w') as f:
        json.dump(best_res, f, indent=4)
    print(f"✅ Saved locally to: {local_file}")

    # 2. Copy to RESULT_DIR using shell command
    # We put quotes "" around the path in case RESULT_DIR has spaces
    !cp {local_file} "{RESULT_DIR}"

    print(f"✅ Copied to: {RESULT_DIR}/{local_file}")
```
